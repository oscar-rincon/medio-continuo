```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
mpl.rcParams['figure.dpi'] = 120
```

```python
L = 2
c = 0.25
q = 50e6

I = c**3/12
nx, ny = (100, 100)
x = np.linspace(-L, L, nx)
y = np.linspace(-c, c, ny)
X, Y = np.meshgrid(x, y)

K = -(q/2/I)

sx = K*(X**2*Y - 2*Y**3/2)
sy = K*(Y**3/3 - c**2*Y + 2*c**3/3)
txy = K*(c**2 - Y**2)*X
```

```python
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(122, projection='3d', proj_type = 'ortho')
ax.view_init(azim=-90, elev=-90)
surf = ax.plot_surface(X, Y, sx, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('$\sigma_{xx}$')
fig.colorbar(surf, shrink=0.2)
ax.set_aspect('equalxy', 'box')
ax._axis3don = False

fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(122, projection='3d', proj_type = 'ortho')
ax.view_init(azim=-90, elev=-90)
surf = ax.plot_surface(X, Y, sy, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('$\sigma_{yy}$')
fig.colorbar(surf, shrink=0.2)
ax.set_aspect('equalxy', 'box')
ax._axis3don = False

fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot(122, projection='3d', proj_type = 'ortho')
ax.view_init(azim=-90, elev=-90)
surf = ax.plot_surface(X, Y, txy, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('$\\tau_{xy}$')
fig.colorbar(surf, shrink=0.2)
ax.set_aspect('equalxy', 'box')
ax._axis3don = False
```

```python
C = 0.5*(sx + sy)
R = np.sqrt( (0.5*(sx - sy))**2 + txy**2)
s1 = C + R
s2 = C - R
```

```python
fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(122, projection='3d', proj_type = 'ortho')
ax.view_init(azim=-90, elev=-90)
surf = ax.plot_surface(X, Y, s1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('$\sigma_{1}$')
fig.colorbar(surf, shrink=0.2)
ax.set_aspect('equalxy', 'box')
ax._axis3don = False

fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(122, projection='3d', proj_type = 'ortho')
ax.view_init(azim=-90, elev=-90)
surf = ax.plot_surface(X, Y, s2, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('$\sigma_{2}$')
fig.colorbar(surf, shrink=0.2)
ax.set_aspect('equalxy', 'box')
ax._axis3don = False

fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(122, projection='3d', proj_type = 'ortho')
ax.view_init(azim=-90, elev=-90)
surf = ax.plot_surface(X, Y, R, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('$\\tau_{max}$')
fig.colorbar(surf, shrink=0.2)
ax.set_aspect('equalxy', 'box')
ax._axis3don = False
```
