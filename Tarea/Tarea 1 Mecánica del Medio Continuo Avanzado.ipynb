## Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'sans'
from matplotlib.ticker import MultipleLocator
import seaborn as sns 
%matplotlib inline
```

**a)** Para el punto A de la barra con coordenadas  $(x,y,z) = (L, 0, H)$ grafique los desplazamientos $u$ y $w$ en función de tiempo.

```python
E1 = 200_000
E2 = 1_000_000
L = 1000
H = 10
t=np.arange(0,501,1)
sigma_x = 4 * t
u=(L/E1)*(sigma_x)
w=-(H*(H**2)/E2)*(sigma_x)
```

```python
# Graficar
plt.figure(figsize=(3.0, 3.0))  # Increase figure size
ax = plt.gca()
sns.lineplot(x=t, y=sigma_x, color="blue", linewidth=1, label="Data")

# Configure labels and titles
ax.set_xlabel(r'Tiempo (s)', fontsize=8)  # Increase font size
ax.set_ylabel(r"$\sigma_{x}$", fontsize=8)  # Increase font size
plt.xticks(np.arange(0, 600, 100), fontsize=8)  # Increase font size
plt.yticks(np.arange(0, 2200, 200), fontsize=8)  # Increase font size
plt.xlim([0, 500])
plt.ylim([0, 2000])
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(100))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Adjust bottom spine width
ax.spines['left'].set_linewidth(0.7)  # Adjust left spine width
ax.tick_params(axis='both', which='both', width=0.7)  # Adjust tick parameters

# Include grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Include grid lines

# Show the plot
plt.tight_layout()
plt.savefig("sigma_x_t.pdf", format="pdf", dpi=300)
plt.show()
```

```python
# Graficar
plt.figure(figsize=(2.5, 2.5))  # Increase figure size
ax = plt.gca()
sns.lineplot(x=t, y=u, color="blue", linewidth=1, label="Data")

# Configure labels and titles
ax.set_xlabel(r'Tiempo (s)', fontsize=8)  # Increase font size
ax.set_ylabel(r"$u$ (cm)", fontsize=8)  # Increase font size
plt.xticks(np.arange(0, 700, 100), fontsize=8)  # Increase font size
plt.yticks(np.arange(0, 11, 1), fontsize=8)  # Increase font size
plt.xlim([0, 500])
plt.ylim([0, 10])
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Adjust bottom spine width
ax.spines['left'].set_linewidth(0.7)  # Adjust left spine width
ax.tick_params(axis='both', which='both', width=0.7)  # Adjust tick parameters
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))
# Include grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Include grid lines

# Show the plot
plt.tight_layout()
plt.savefig("u_t.pdf", format="pdf", dpi=300)
plt.show()
```

**c)** Si el valor de $\sigma_x = 50 \text{ kgf/cm}^2$ y se estudian solo los puntos de la cara $z = 1.0$ cm determinar el valor de la distorsión angular máxima y las coordenadas $x,y,z$ de los puntos  en que se presenta. 

```python
z=1 #cm
sig_x=50#kgf/cm2
x=np.arange(-1000,1000,10) #cm
y=np.arange(-10,10,0.1) #cm
w_yz = -y*z*sig_x/E2 * 10000
```

```python
plt.figure(figsize=(2.8, 2.5))  # Increase figure size
ax = plt.gca()
sns.lineplot(x=y, y=w_yz, color="blue", linewidth=1, label="Data")
# Configure labels and titles
ax.set_xlabel(r'$y$ (cm)', fontsize=8)  # Increase font size
ax.set_ylabel(r"$w \times 10^4$", fontsize=8)  # Increase font size
plt.xticks(np.arange(-10, 11, 2), fontsize=8)  # Increase font size
plt.yticks(np.arange(-5, 6, 2), fontsize=8)  # Increase font size
plt.xlim([-10, 10])
plt.ylim([-5, 5])
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(1))
ax.legend().set_visible(False)
sns.despine()
ax.spines['bottom'].set_linewidth(0.7)  # Adjust bottom spine width
ax.spines['left'].set_linewidth(0.7)  # Adjust left spine width
ax.tick_params(axis='both', which='both', width=0.7)  # Adjust tick parameters
# Include grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Include grid lines
# Show the plot
plt.tight_layout()
plt.savefig("w_yz.pdf", format="pdf", dpi=300)
plt.show()
```

```python
# Calculate w_yz
X, Y = np.meshgrid(x, y)
w_yz = -Y * z * sig_x / E2 * 10000
# Plotting
plt.figure(figsize=(3.2, 2.5))  # Increase figure size
ax = plt.gca()
plt.pcolor(X, Y, w_yz, cmap='gray')
plt.xlabel(r'$x$ (cm)', fontsize=8)
plt.ylabel(r'$y$ (cm)', fontsize=8)
plt.ylim([-10, 10])
plt.xlim([-1000, 1000])
ax.xaxis.set_minor_locator(MultipleLocator(200))
ax.yaxis.set_minor_locator(MultipleLocator(1))
plt.xticks(np.arange(-1000, 1100, 400), fontsize=8)  # Increase font size
plt.yticks(np.arange(-10, 11, 2), fontsize=8)  # Increase font size
plt.colorbar(label=(r"$w_{yz} \times 10^4$"))  # Set colorbar fontsize to 8pt
plt.tick_params(labelsize=8)  # Set fontsize of colorbar ticks
# Show the plot
plt.tight_layout()
plt.savefig("w_yz.pdf", format="pdf", dpi=100)
plt.show()
```
